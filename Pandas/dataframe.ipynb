#### Filtering DataFrame

```python
ipl.head(2)
```

```python
# find all the final winner
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
# how many super over finishes have occurred
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has csk won in kolkata
ipl[(ipl['WinningTeam'] == 'Chennai Super Kings')&(ipl['City'] =='Kolkata')].shape[0]
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and vote > 10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
# action movies with rating higher than 7.5
# movies[(movies['genres'].str.split('|').apply(lambda x:'Action' in x)) & (movies['imdb_rating'] > 7.5)].shape[0]
movies[(movies['genres'].str.contains('Action')) & (movies['imdb_rating'] > 7.5)].shape[0]
```

```python
# write a function that can return the track record of 2 teams against each other
def trackRecord(team1,team2):
    r = (ipl['Team1'] == team1) & (ipl['Team2'] == team2)
    return ipl[r]
trackRecord('Chennai Super Kings','Mumbai Indians')
```
