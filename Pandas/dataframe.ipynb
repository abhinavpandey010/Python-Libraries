#### Creating Dataframe

```python
# using lists
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]
pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
# using dictionary
student_dict = {
    'name':['nitish','ankit','rupesh','amit'],
    'iq':[100,90,120,80],
    'marks':[80,70,100,50],
    'package':[10,7,14,2]
}
students = pd.DataFrame(student_dict)
students.set_index('name',inplace=True)
students
```
