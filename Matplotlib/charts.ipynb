#### Bar charts

- Catogories comparison for data analysis
- bar charts can be vertical and horizontal
- dont use with continous data

#### Pie chrts

- Proportional representation 
- Whole Representation

#### Histograms

- Numerical distribution
- Data Insights
- use with continous data


```python
import matplotlib.pyplot as plt 
# plt.bar(x, height, color = '',width = , label - '')

product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.bar(product,sales,color = 'orange', label = 'sales 2025')
plt.xlabel('product')
plt.ylabel(sales)
plt.title('Product sale comparison')
plt.legend()
```

```python
# plt.pie(values,labels = label_list, color = color_list, autopct = '%1.1f%%') 

regions = ['North','East','West','South']
revenue = [3000,2000,1500,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%', colors = ['gold','skyblue','seagreen','coral'])
plt.title('Revenue contribution')
```

```python
# plt.hist(data, bins = number_of_bins, color = 'color name', edgecolor= '')

scores = [40,56,23,89,67,98,34,23,87,78,58]

plt.hist(scores, bins = 5, color = 'purple', edgecolor= 'black')
plt.xlabel('Score range')
plt.ylabel('No of students')
plt.title('Score distribution of students')
```
