### Key terms 

1. Datapoint - x-axis,y-axis
2. Figure - Entire canvas or paper,whole window 
3. Axes - Graph box(multiple graphs can be built on single figure)
4. Plot - Actual line,bar and shapes that visible on graph
5. Marker - Symbol(o[circle],s[square],d[dot]) that represent data points and used to show data variation
6. Line Style - type of line(dash line,single line)
7. Legend - Small box that show that line and color what represents
8. Grid - horizontal/vertical line on graph
9. DPI - dots per inch
