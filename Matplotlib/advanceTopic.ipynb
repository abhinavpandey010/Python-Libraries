```python
import numpy as np
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
import pandas as pd
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df["target"] = iris.target
print(df)
```

```python

df.head()
```

#### Scatter plot more functions

```python
plt.figure(figsize=(10,5))# can adjust size of graph
# c = can give color to results by assigning result column 
# cmap = changes color of points 
# alpha = control transparency of points
plt.scatter(df["sepal length (cm)"],df["petal length (cm)"],c=df["target"],cmap="cool",alpha=0.2)
plt.xlabel("sepal length (cm)")
plt.ylabel("petal length (cm)")
# generate color bar to give insights(range) of points
plt.colorbar()
```

##### Annottations

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
for i in range(len(x)):
    plt.text(x[i],y[i],f'point {i+1}',fontdict={'size':12,'color':'orange'})
```

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
sampledf = batters.head(100).sample(25,random_state=5)
```

```python
plt.figure(figsize=(20,15))
# s = can control size of scatter points by passing a list
plt.scatter(sampledf['avg'],sampledf['strike_rate'],s=sampledf['runs'])

# plot a line have strike rate of 130 that divide graph on basis of strile rate
# can plot multiple lines 
plt.axhline(130,color='red')
# this dive line on tha basis of average of player
plt.axvline(30,color='red')


for i in range(len(sampledf)):
    plt.text(sampledf['avg'].values[i],sampledf['strike_rate'].values[i],sampledf['batter'].values[i])
```
