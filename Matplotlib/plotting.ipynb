#### plot method


- plt.plot(x,y,color='color name', linestyle =value,linewidth='',marker='',label = '')

```python
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000, 1500, 1200, 1800]

plt.plot(months,sales, color = 'blue', linestyle = '--',linewidth = 2,marker = 'o',label = '2025 sales data')

plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly sales data report')
plt.legend(loc='lower right') # by default it set legend to upper right # its work only when label is set
plt.grid(color = 'grey', linestyle = ':',linewidth = 1)

plt.xlim(1,4) # by this method can control range of x-axis
plt.ylim(0,2000) # same working of y-axis limit

plt.xticks([1,2,3,4],['M1','M2','M3','M4']) # replaces x-axis data points
# plt.yticks() 

```

#### Scatter plot

```python
import matplotlib.pyplot as plt
hour_studies = [1,2,3,4,5,6,7,8]
exam_scores = [55,60,65,70,85,89,75,69]


plt.scatter(hour_studies,exam_scores,color = 'green', marker = 'o', label = 'student data')

plt.xlabel("Hour studied")
plt.ylabel("Exam Score")

plt.title("Relationship between student exam scores and hour_studies")
plt.legend()
```
