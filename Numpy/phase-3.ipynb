## Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # paradise biryani
    [2, 120000, 140000, 160000, 190000], # beiging bites
    [3, 200000, 230000, 260000, 300000], # pizza hub
    [4, 180000, 210000, 240000, 270000], # burger point
    [5, 160000, 185000, 205000, 230000]  # chai point
])

print('==== Zomato sales analysis ====')
print('\n Sale data shape', sales_data.shape)
print('\n Sample data for 1st 3 restau\n', sales_data[:3])
print('\n Sample data for 1st 3 restau exclude index\n', sales_data[:3,1:])
```

```python
# total sales per year
print(np.sum(sales_data,axis = 0))
yearly_total = np.sum(sales_data[:,1:], axis=0)
print(yearly_total)

# minimum sales per restaurant  

min_sales = np.min(sales_data[:,1:], axis = 1)
print(min_sales)

# max sales per year
max_sales = np.max(sales_data[:,1:],axis = 0)
print(max_sales)

# average sales per restaurant
avg_sales = np.mean(sales_data[:,1:], axis = 1)
print(avg_sales)
```

```python
# cumulative sum

cumsum = np.cumsum(sales_data[:,1:], axis = 1)
print(cumsum)

plt.Figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title('Average cumulative sales per year across all restaurant ')
plt.xlabel('year')
plt.ylabel('sales')
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4])
vector2 = np.array([5,6,7,8])
print('vector addition: ', vector1 + vector2 )
print('vector multiplication: ', vector1 * vector2)
print('dot product: ', np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print('angle: ', angle)
```
